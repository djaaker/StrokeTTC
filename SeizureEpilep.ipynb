```python
import pandas as pd
import re

# Assuming abnBEX is already defined
# Define the regex pattern
pattern = r'(?<=/edf/\d{3}/)([a-z]+?)/'

# Apply the regex to extract the names and store them in a list, ensuring that only string values are processed
abnAll = abnBEX['path_new'].apply(lambda x: re.search(pattern, x).group(1) if isinstance(x, str) and re.search(pattern, x) else None).dropna().tolist()
normAll = normBEX['path_new'].apply(lambda x: re.search(pattern, x).group(1) if isinstance(x, str) and re.search(pattern, x) else None).dropna().tolist()
print(len(abnAll))
```

```python
# Assuming abnAll, seizureAll, and epilepsyAll are already defined as lists of strings

# Assuming epilepsyAll and seizureAll are defined as lists of strings

# Remove all '/' characters from the strings in epilepsyAll and seizureAll
epilepsyAll = [entry.replace('/', '') for entry in epilepsyAll]
seizureAll = [entry.replace('/', '') for entry in seizureAll]

# Print to verify the result (optional)
print("epilepsyAll:", epilepsyAll)
print("seizureAll:", seizureAll)
print("normAll:", normAll)

# Convert lists to sets
set_abnAll = set(abnAll)
set_seizureAll = set(seizureAll)
set_epilepsyAll = set(epilepsyAll)
set_normAll = set(normAll)

print(len(set_abnAll))
print(len(set_seizureAll))
print(len(set_epilepsyAll))

# Find common elements between abnAll and seizureAll
common_abn_seizure = set_abnAll.intersection(set_seizureAll)

# Find common elements between abnAll and epilepsyAll
common_abn_epilepsy = set_abnAll.intersection(set_epilepsyAll)

# Print the number of common elements
print("Number of elements in abnAll also in seizureAll:", len(common_abn_seizure))
print("Number of elements in abnAll also in epilepsyAll:", len(common_abn_epilepsy))

# Assuming `common_all_three` contains the common elements in all three lists

# Calculate the number of elements not in common by subtracting the length of common elements from the total number of unique elements in all three lists
all_elements = set(abnAll) | set(seizureAll) | set(epilepsyAll) | set(normAll) # Union of all elements
```

```python
def extract_lines_start_a_end_underscore(file_path):
    results = []
    
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith('a') and line.endswith('_'):
                results.append(line)
    
    return results

# File path provided by user
file_path = r"C:\Users\dalto\Box Sync\New Text Document.txt"
# Calling the function and storing the result
results = extract_lines_start_a_end_underscore(file_path)

# Print the results
for result in results:
    print(result)
```
